```python
import pandas as pd
import numpy as np
```

```python
file_path = ("C:\\Users\\Priyanka\\Downloads\\Normal_variant.csv")
```

```python
data = pd.read_csv(file_path)
```

```python
data.head()
```

data.head()

```python
# Step 1: Convert 'Normal_SM_AF' to numeric if needed (handling cases with multiple values)
data['Normal_SM_AF'] = data['Normal_SM_AF'].apply(lambda x: float(x.split(',')[0]) if isinstance(x, str) else x)

# Step 2: Calculate the median background mutation level from normal tissue
background_mutation_level = data['Normal_SM_AF'].median()
```

```python
data.head()
```

```python
background_mutation_level
```

```python
# Step 3: Ensure read depth ('Normal_SM_DP') is numeric
data['Normal_SM_DP'] = pd.to_numeric(data['Normal_SM_DP'], errors='coerce')  # Handle non-numeric values
```

```python
# Step 4: Calculate RPM (Reads Per Million)
total_read_depth = data['Normal_SM_DP'].sum()  # Total read depth across all samples
data['Normal_SM_RPM'] = (data['Normal_SM_DP'] / total_read_depth) * 1e6  # Calculate RPM
```

```python
data
```

```python
# Step 5: Define a threshold for confident somatic mutation calling
# We will use the background mutation level from normal tissue and adjust it for somatic mutation calling
confidence_threshold_AF = background_mutation_level * 1.0  # Set allele frequency to 1x the background level for somatic mutations
confidence_threshold_RPM = 1000  # Set a fixed RPM threshold (can be adjusted based on experiment)
```

```python
confidence_threshold_AF
```

```python
confidence_threshold_RPM
```

```python
# Step 6: Identify confidently called somatic mutations based on both AF and RPM
data['Confident_Somatic_Mutation'] = (
    (data['Normal_SM_AF'] > confidence_threshold_AF) & 
    (data['Normal_SM_RPM'] > confidence_threshold_RPM)  # RPM threshold for sufficient coverage
)
```

```python
# Step 7: Display the results
print("\nData with Confident Somatic Mutation Status and RPM:")
print(data[['CHROM', 'POS', 'REF', 'ALT', 'Normal_SM_AF', 'Normal_SM_DP', 'Normal_SM_RPM', 'Confident_Somatic_Mutation']])
```
